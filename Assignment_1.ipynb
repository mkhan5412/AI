```
df = df_Application_train.sample(frac=.40)
```

```
df
```

```
category_columns = df.select_dtypes(include='object').columns
```

```
integer_columns = df.select_dtypes(include=['int64','float']).columns
```

```
for i in category_columns:
  df[i].fillna(df[i].mode()[0], inplace=True )
```

```
for i in integer_columns:
  df[i].fillna(df[i].mean(), inplace=True)
```

```
X=df.drop(['TARGET','SK_ID_CURR'],axis=1)
Y=df['TARGET']
```

```
X.shape
```

```
dictionary={'Higher education':4, 'Secondary / secondary special':1,
       'Incomplete higher':3, 'Lower secondary':0, 'Academic degree':2}
X['NAME_EDUCATION_TYPE']=X['NAME_EDUCATION_TYPE'].map(dictionary)
```

```
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for i in category_columns:
  if i!='NAME_EDUCATION_TYPE':
    X[i]=le.fit_transform(X[i])
```

```
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=123)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

classifier=LogisticRegression(max_iter=1000)
classifier.fit(X_train,Y_train)
pred=classifier.predict(X_test)
print(classification_report(pred,Y_test))
```

```
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier().fit(X_train,Y_train)
predictions=rf.predict(X_test)
print(classification_report(predictions,Y_test))
```

```

from xgboost import XGBClassifier

xgb = XGBClassifier(objective='binary:logistic')
xgb.fit(X_train,Y_train)
predictions=xgb.predict(X_test)
print(classification_report(predictions,Y_test))
```

```
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, Y_train)
predictions = nb.predict(X_test)

print(classification_report(predictions, Y_test))
```
